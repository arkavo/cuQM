```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
X = np.linspace(0,2*np.pi,200)

Y_sin = np.sin(X)
Y_cos = np.cos(X)
```

```python
sinusoid_data = pd.DataFrame()

for j in range(60000):
    L = np.random.randint(10)
    Coeffs = np.random.randint(10,size=[10])
    Powers = np.random.randint(2,size=[10])
    print(L,Coeffs,Powers)
    Y = np.zeros_like(X)
    for i in range(10):
        Y += Coeffs[i]*np.power(Y_cos,i)*np.power(Y_sin,10-i)*Powers[i]

    sinusoid_data.append(pd.DataFrame([Coeffs,Powers,Y]))
```
